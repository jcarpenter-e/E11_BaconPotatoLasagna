```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
pm25data = pd.read_csv("pm25_data.csv")
pm25data
```

```python
plt.hist(pm25data["Particles > 0.3um / 0.1L air:"],bins=40,alpha = 0.5, color = 'slategrey')
plt.hist(pm25data["Particles > 0.5um / 0.1L air:"],bins=30,alpha = 0.5, color = 'pink')
plt.hist(pm25data["Particles > 2.5um / 0.1L air:"],bins=15,alpha = 0.25, color = 'm')
plt.hist(pm25data["Particles > 1.0um / 0.1L air:"],bins=10,alpha = 0.5, color = 'aqua')
plt.hist(pm25data["Particles > 5.0um / 0.1L air:"],bins=10,alpha = 0.5, color = 'gold')
plt.hist(pm25data["Particles > 10 um / 0.1L air:"],bins=10,alpha = 0.5, color = 'orange')
plt.ylim(top=35)
# Here, we are measuring the frequency of particles that are greater than a certain size, while there are many particles greater than 0.3 micrometers
# as the sensor sorts for higher and higher sizes, the counts get increasingly smaller. On the histogram, it almost looks like the data is getting
# increasing condensed to the left
```
