```python
import matplotlib.pyplot as plt
import numpy as np
import pandas
import datetime
```

```python
filedata = pandas.read_csv("Loc1_session1_group1_2025-04-16_15-23-08.csv")
Time = filedata["Time"]
Lat = filedata["Latitude"]
Lon = filedata["Longitude"]
Count = filedata["counts"]
```

```python
plt.plot(Time,Count,color="maroon")
plt.title("Time vs Radiation")
plt.xlabel("Time (Seconds)")
plt.ylabel("Radiation (Counts per Second)")
```

```python
Distance = np.sqrt(Lat**2 + Lon**2)
plt.scatter(Distance,Count,color="maroon")
plt.title("Distance vs Radiation")
plt.xlabel("Distance (Degrees from Start)")
plt.ylabel("Radiation (Counts per Second)")
plt.ylim(0,60)
```

```python
plt.hist(Count,bins=20,color="maroon")
plt.title("Counts per Second")
plt.xlabel("Counts")
plt.ylabel("Frequency")
```

```python
### Mean of Count Frequency
print(f" The mean Count frequency is {round( np.mean(Count), 3 )}")
print(f" The standard deviation is {round( np.std(Count), 3 )}")
print("--> The data set does approximately follow a normal distribution."
```

Any collected latitude/longitude/distance values wont follow a normal dist. since it depends on where we walked.

Humidity and pressure should be roughly normal since the outside value is nearly constant. 

Altitude won't be normal since it also depends on where we walk.

Temperature might not be normal since it is affected by other factors, like indoor/outdoor, nearby ventilation, etc. 

```python
print("According to statistics, the standard deviation should be approximately sqrt(Var(X)), so we would expect")
print(f"a {round(np.std(Count)**2, 3 )} frequency variance from the mean. That is relatively similar to the recorded")
print(f"variance of {max(Count)-min(Count)}.")
```

```python
Uncertainty = np.sqrt(len(Count)) / np.sqrt(2)
print(f"Our general uncertainty is {round(Uncertainty,3)}, so our measured variance is within the range of resonable uncertainty.")
```
