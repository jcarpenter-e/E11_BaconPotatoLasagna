```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
InsideDataRaw = pd.read_csv("insidedata.csv")
InsideData = InsideDataRaw.drop(0)
InsideData
```

```python
OutsideDataRaw = pd.read_csv("outsidedata.csv")
OutsideData = OutsideDataRaw.drop(0)
OutsideData
```

```python
fig, linplt = plt.subplots(4, 2, figsize=(20, 15))
linplt[0,0].hist(OutsideData["Temperature"],alpha = 0.5, bins = 50)
linplt[0,0].hist(InsideData["Temperature"], alpha = 0.5, bins=50)
linplt[0,0].set_title("Temperature")

linplt[0,1].hist(OutsideData["Gas"], alpha = 0.5, bins = 50)
linplt[0,1].hist(InsideData["Gas"], alpha = 0.5, bins = 50)
linplt[0,1].set_title("Gas")

linplt[1,0].hist(OutsideData["Humidity"], alpha = 0.5, bins = 40)
linplt[1,0].hist(InsideData["Humidity"], alpha = 0.5, bins = 40)
linplt[1,0].set_title("Humidity")

linplt[1,1].hist(OutsideData["Pressure"], alpha = 0.5, bins = 10)
linplt[1,1].hist(InsideData["Pressure"], alpha = 0.5, bins = 10)
linplt[1,1].set_title("Pressure")

linplt[2,0].hist(InsideData["PM 1.0 Standard Conc."],alpha = 0.5, bins = 3)
linplt[2,0].hist(OutsideData["PM 1.0 Standard Conc."],alpha = 0.5, bins = 3)
linplt[2,0].set_title("PM 1.0")

linplt[2,1].hist(OutsideData["PM 2.5 Standard Conc."] ,alpha = 0.5, bins = 5)
linplt[2,1].hist(InsideData["PM 2.5 Standard Conc."],alpha = 0.5, bins = 5)
linplt[2,1].set_title("PM 2.5")

linplt[3,0].hist(InsideData["PM 10 Standard Conc."],alpha = 0.5, bins = 5)
linplt[3,0].hist(OutsideData["PM 10 Standard Conc."],alpha = 0.5, bins = 5)
linplt[3,0].set_title("PM 10")
```

```python
Omean = np.mean(OutsideData,0)
Imean = np.mean(InsideData,0)
Omeanstd = np.std(OutsideData,0)/np.sqrt(299)
Imeanstd = np.std(InsideData,0)/np.sqrt(299)
Imeanstd
```

```python
statisticsig1 = (Imean > (Omean + 3*Omeanstd)) | (Imean < (Omean - 3*Omeanstd))
statisticsig2 = (Omean > (Imean + 3*Imeanstd)) | (Omean < (Imean - 3*Imeanstd))
statisticsig1, statisticsig2
```

#### Answers for 1.0.1
Our mean uncertainity can be found by taking the experimental mean, and dividing my the square root of the amount of measurements we took, in this case 299 (we dropped the first 1). By seeing how many of our data sets lie outside 3 STDs of the others, we can see that for all data points, we are sampling from two, statistically different, distributions with less than a 1% chance that they are similar. We also conclude that although most of the data does not resemble that of a Gaussian, we must not have taken enough measurements for the Central Limit Theorem to occur, though it is true for all N, not just infinity.

The back correlation between the values represents the Kendalls Tau measure, given that the correlation is imperfect but generally correlated, with some observable points of concordance and discordance.
