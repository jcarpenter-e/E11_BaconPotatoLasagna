```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
combineddata = pd.read_csv("BME680_PM25.csv")
cleaned = combineddata.drop(0)
cleaned
```

```python
fig, linplt = plt.subplots(2, 3, figsize=(20, 10))
linplt[0,0].plot(cleaned["Time"],cleaned["Temperature"])
linplt[0,0].set_title("Temperature")

linplt[0,1].plot(cleaned["Time"],cleaned["Gas"])
linplt[0,1].set_title("Gas")

linplt[0,2].plot(cleaned["Time"],cleaned["Humidity"])
linplt[0,2].set_title("Humidity")

linplt[1,0].plot(cleaned["Time"],cleaned["Pressure"])
linplt[1,0].set_title("Pressure")

linplt[1,1].plot(cleaned["Time"],cleaned["Altitude"])
linplt[1,1].set_title("Altitude")

linplt[1,2].plot(cleaned["Time"],cleaned["Particles > 1.0um / 0.1L air:"], label = ">1um")
linplt[1,2].plot(cleaned["Time"],cleaned["Particles > 2.5um / 0.1L air:"], label = ">2.5um")
linplt[1,2].plot(cleaned["Time"],cleaned["Particles > 0.3um / 0.1L air:"], label = ">0.3um")
linplt[1,2].plot(cleaned["Time"],cleaned["Particles > 0.5um / 0.1L air:"], label = ">0.5um")
linplt[1,2].plot(cleaned["Time"],cleaned["Particles > 5.0um / 0.1L air:"], label = ">5um")
linplt[1,2].plot(cleaned["Time"],cleaned["Particles > 10 um / 0.1L air:"], label = ">10um")
linplt[1,2].legend(fontsize='small')
linplt[1,2].set_title("Particles per 1L air")
```
