```python
def Histogram(NAME): # Probability of each Count Value
    data = pandas.read_csv(NAME)
    n = data["Counts in the Last Minute"]
    dsum = pandas.DataFrame.sum(data)
    mean = dsum / len(data)
    plt.hist(n,bins=10,alpha=0.5,color=(colorcount,1-colorcount,1))
    return mean
```

```python
def Mean_SD(NAME):
    data = pandas.read_csv(NAME)
    mean = pandas.DataFrame.mean(data)
    sd = pandas.DataFrame.std(data)
    return mean,sd
```

```python
def Sum_Mean_SD(NAME):
    data = pandas.read_csv(NAME)
    sum = pandas.DataFrame.sum(data)
    mean = pandas.DataFrame.mean(data)
    sd = pandas.DataFrame.std(data)
    return data, sum, mean, sd
```
